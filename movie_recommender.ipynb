```python
import numpy as np
import pandas as pd
import json
# https://www.kaggle.com/rounakbanik/the-movies-dataset 캐글 The Movies Dataset

df = pd.read_csv('movies_metadata.csv')
df.head()
```

```python
df = df[['id', 'original_title', 'original_language','genres']] # 필요한 칼럼 분류
df = df.rename(columns={'id':'movieId'}, ascending=False)
df = df[df['original_language'=='en']] # 대부분의 언어 분류
```

```python
df.head() # 추출한 데이터 보관
```

```python
ratings = pd.read_csv('ratings_small.csv') # 평가 데이터 
ratings = ratings[['userId','movieId','rating']] 
ratings.head()
```

```python
ratings.describe() # 데이터의 대략적 개요
```

```python
'''
Refine Dataset
1.df와 ratings의 movieId를 전부 숫자로 바꿔서 비교
2.genres  배열형태로 가공
'''
df.movieId = pd.to_numeric(df.movieId, errors='coerce') # pd.to_numeric 문자열을 숫자로 변환
ratings.movieId = pd.to_numeric(ratings.movieId, errors= 'coerce')

```

```python
def parse_genres(genres_str):
    genres = json.loads(genres_str.replace('\'', '"')) # json.loads사용 + replace따옴표
    
    genres_list = []
    for g in genres:
        genres_list.append(g['name'])

    return genres_list #genres_list에 genres['name']만 추가하는 리스트

df['genres'] = df['genres'].apply(parse_genres) # df.apply()각 행에 해당 함수 적용

df.head()
```

```python
'''
Merge df and Ratings 
한 영화에 대한 다중 사용자들에 대한 평가 데이터셋이 생성됨
'''
data = pd.merge(ratings, df, on = 'movieId', how='inner') # pd.merge, inner 사용
data.head()
```

```python
'''
Pivor Table
data.pivot_table(index='', columns='', values='')
'''
table = data.pivot_table(index='userId', columns= 'original_title', values = 'rating')
table.head(20)
```

```python
'''
피어슨 상관관계를 선형적으로 표현
Pearson correlation
https://namu.wiki/w/%EC%83%81%EA%B4%80%EA%B3%84%EC%88%98#s-2
'''
GENRE_WEIGHT = 0.1

def personR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c**2))
#상관관계는 분모가 클 수록 o, 분자가 클 수록 x

def recommend(input_movie, table, n, similar_genre=True):
    # 입력 받은 영화 장르 저장
    input_genres = df[df['original_title'] == input_movie]['genres'].iloc(0)[0]
    
    result = []
    for title in table.columns:
        if title == input_movie:
            continue # 똑같은 영화 추천시 건너뜀
            
        #rating comparison
        cor = personR(table[input_movie], table[title]) # 상관관계 반환 ,값이 높을 수록 o
        
        #genre comparison
        if similar_genre and len(input_genres) > 0:
            temp_genres = df[df['original_title'] == title]['genres'].iloc(0)[0]
            #비교 중인 영화 장르와 입력받은 장르 비교 np.isin() 배열 비교해서 같은 요소 있으면 True 반환
            same_count = np.sum(np.isin(input_genres, temp_genres)) # 
            cor += (GENRE_WEIGHT * same_count) # 같은 요소가 있을수록 장르에 가중치,  최종반환
            
        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_genres))
            
    result.sort(key=lambda r: r[1], reverse=True) # ratings높은 순서대로 내림차순 정렬

    return result[:n] #n개 까지만 반환해줌
    
# table 에서 input_movie와 유사한 값 10개까지 추출
recommend_result = recommend('올드보이', table, 10, similar_genre=True)

pd.DataFrame(recommend_result, columns = ['추천 영화 제목', '추천점수', '장르'])
```
